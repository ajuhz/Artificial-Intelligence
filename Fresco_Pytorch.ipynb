```
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
img = np.array(Image.open("/content/gdrive/My Drive/Python/bird.jpeg").resize((224,224)))
print(img.shape)
img_tensor = torch.Tensor(img)
plt.imshow(img)
print("shape of image tensor {}".format(img_tensor.shape))
```

```
# getting only one channel of the image
channen_0 = img_tensor[:, : , 0]
print("shape of first channel of image data:{}".format(channen_0.shape))
plt.imshow(channen_0)
```

```
# slicing part of the image
channen_0 = img_tensor[100:200, : , 0]
print("shape of first channel of image data:{}".format(channen_0.shape))
plt.imshow(channen_0)
```

```
# reshaping tensor
x=torch.Tensor([[1,2,3],[5,6,7]])
print('Before reshping tensor {}'.format(x.shape))
print(x)
x=x.view(-1,1)
print('After reshaping the tensor {}'.format(x.shape))
print(x)
```

```
#Useful operations
#topk
x=torch.Tensor([[2,3,4],[4,5,6]])
x.topk(k=2,dim=1)
```

```
#torch.tensor.item() to convert one value to scaler
x[0][1].item()
```

```
# inplace operation 
# in Pytorch any method followed by  _ will do inplace opperation
print('Before in place operation value of x',x)
x.add_(2)
print('After in place operation value of x',x)
```
